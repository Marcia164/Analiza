```python
def fizzbuzz(n, x = 3, y = 5):

    if n % x == 0 and n % y != 0 :
        return "Fizz"
    elif n % y == 0 and n % x != 0:
        return "Buzz"
    elif n % x == 0 and n % y == 0:
        return "FizzBuzz"
    else:
        return ""

fizzbuzz(9)
```

```python
def contains_letter(letter: str, word: str) -> str:
    return "Yes" if letter in word else "No"
contains_letter("5", "123456")
```

```python
def znajdz_samogloski(word : str) -> list:
    lista = []
    lista.extend(a for a in word if a in ["a", "e", "y", "u", "i", "o"])
    return lista
znajdz_samogloski("")
```
